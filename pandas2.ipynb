```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.rand(5,6), 
                  ['a', 'b', 'c', 'd', 'e'], 
                  ['U', 'V', 'W', 'X', 'Y', 'Z'])
df
```

```python
table = df[df>0.15] #reiksmes NAN reiskia, kad siu reiksmiu nera lenteleje. NAN not a number
table
```

```python
table.isnull() #grąžina lentelę, kur true reikšmės yra NaN t.y skaicius yra didesnis uz 0.15
```

```python
table.isnull().sum() # Rodo kiek kuriame stulpelyje NaN reikšmių
```

```python
table['Y'].isnull() # Galima tikrinti atskiruose stulpeliuose
```

```python
table['V'].isnull().sum() #suskaiciuoja kiek stulpelyje yra NAN reiksmiu
```

```python
#pirmas metodas tvarkytis su trūkstamais duomenimis yra .dropna(). Jis išmeta visas eilutes, kuriose yra bent viena NaN reikšmė:
table.dropna()
```

```python
#jeigu norime išmesti stulpelius, turinčius bent vieną NaN, turime nurodyti ašį:
table.dropna(axis=1) #axis = 1 stulpeliai (axis = 0 eilutes, nurodyti nebutina, nes yra pagal nutylejima)
```

```python
#Galime išmesti tik tas eilutes, kurie turi mažiau, negu (pvz) 5 NE NaN reikšmes:
table.dropna(thresh=5) #minimalus patenkinantis salyga reiksmiu kiekis
```

```python
#Papildomai nurodžius ašį, tą patį galima padaryti ir su stulpeliais.
table.dropna(thresh=4, axis=1)
```

```python
#Jei norime NaN reikšmes pakeisti kažkuo kitu, naudosime .fillna() metodą:
table.fillna('bupkis')
```

```python
#Daugiau prasmės būtų pakeisti pvz. stulpelio vidurkiu, taigi NAN reiksme buvo pakeista Y stulpelio vidurkiu:
table['Y'].fillna(value=table['Y'].mean())
```

```python
table.fillna(value=table.mean()) #visos trukstamos reiksmes buvo pakeistos stulpelio vidurkiu
```

```python
table.loc["a"]
```

```python
table.loc["b"].fillna(value=table.loc['b'].mean()) #uzpildo eilutes vidurkiu NAN reiksmes
```

```python
#Grupavimas

duomenys = {'Šalis': ['Lietuva',
  'Lietuva',
  'Lietuva',
  'Latvija',
  'Latvija',
  'Latvija',
  'Estija',
  'Estija',
  'Estija'],
 'Miestas': ['Vilnius',
  'Kaunas',
  'Klaipėda',
  'Ryga',
  'Ventspilis',
  'Daugpilis',
  'Talinas',
  'Tartu',
  'Pernu'],
 'Gyv': [541, 287, 147, 716, 43, 105, 400, 101, 46]}
data = pd.DataFrame(duomenys)
data
```

```python
#panaudokime .groupby() metodą, viduje nurodydami stulpelį, pagal kurį grupuosime.
data.groupby('Šalis')
```

```python
baltic = data.groupby('Šalis')
baltic
```

```python
#.mean() skaičiuos vidurkį, tam kad isspausdintu reikia nurodyti pvz vidurki, arba max ir kt. sum count
baltic.mean(numeric_only=True) #numeric_only duoda tik duomenis su reiksmemis
```

```python
data.groupby('Šalis').sum(numeric_only=True)
```

```python
# iš groupby objekto galime traukti reikšmes tokiu būdu:
# .count() suskaičiuos kiek yra įrašų grupės stulpeliuose:
baltic.count()
```

```python
baltic.max() # max arciau z min arciau a
#isrusiavo salis pagal abecele, kad visai nera tikslu ir prasminga. Taigi max ir min nėra labai praktiški su string tipo reikšmėmis.
```

```python
#iš groupby objektų galime traukti reikšmes:
baltic.sum(numeric_only=True).loc['Lietuva']
```

```python
#.describe() metodas duoda pagrindinę informaciją apie lentelės duomenis:
baltic.describe()
```

```python
#galime sukeisti stulpelius su eilutėmis su .transpose():
baltic.describe().transpose()
```

```python
baltic.describe().transpose()['Estija']
```

```python
#DF Jungimas
data_main = data[:6] #iki 6
data_main
```

```python
data_bottom = data[6:] #nuo 6
data_bottom
```

```python
#norėdami sujungti data_main su data_bottom naudosime .concat() metodą, viduje įrašydami lentelių, kurias jungsime sąrašą:
pd.concat([data_main, data_bottom])
```

```python
#galime prijungti lentelę iš dešinės:
pd.concat([data_main, data_bottom], axis=1)
```

```python
#pertvarkome duomenis jungimui, data_bottom pervadiname i data_right ir nuresetiname indeksa, pridedami nauja eilute, o sena reiks ismesti
data_right = data_bottom.reset_index()
data_right
```

```python
data_right.drop(columns='index', inplace=True) #ismetam sena indeksa
data_right
```

```python
pd.concat([data_main, data_right], axis=1) #prie pagrindines lenteles jungiame sutavarkyta is desines
```

```python
lenkija = {'Šalis': ['Lenkija', 'Lenkija', 'Lenkija'], 
           'Miestas':['Varšuva', 'Vroclavas', 'Gdanskas'], 
          'Gyv': [1688, 638, 461]}
pl = pd.DataFrame(lenkija)
data_right1 = pd.concat([data_right, pl])
data_right1
#kadangi tvarkingam prijungimui truksta duomenu  pridedame duomenu
```

```python
data_right = data_right1.reset_index().drop(columns='index') #tada prijungtos lenteles reikia sutvarkyti indexa, ji ismesti
data_right
```

```python
pd.concat([data_main, data_right], axis=1)
```

```python
#.merge() Jeigu data_main ir data_right turėtų po vieną sutampantį stulpelį, pvz.:
#susikuriame netikrus naujus parametrus:
key = 'A B C D E F'.split()
data_main['key'] = key
data_right['key'] = key
```

```python
pd.merge(data_main, data_right, on="key") #on= nurodo per kuria vieta jungiame, siuo atveju data_main ir data-right key reiskme sutampa
```

```python
#papildykime pavyzdį dar vienu stulpeliu:
key2 = 'B C D E F G'.split()
data_main['key2'] = key2
data_right['key2'] = key2
```

```python
pd.merge(data_main, data_right, on=['key', 'key2'])
```

```python
#kaip jungti kai jungimo - siuo atveju "key" reiksmes nesutampa
keyX = 'W C D E F W'.split()
data_right['key2'] = keyX
data_right
```

```python
data_main
```

```python
pd.merge(data_main, data_right, on=['key2'])  #jungia tik tas reiksmes kurios sutampa
```

```python
#Yra ir metodas .join(). Jis daro tą patį, ką ir merge, tik indekso pagrindu. Pvz.:
data_main.join(data_right, lsuffix='L', rsuffix='R')
```
