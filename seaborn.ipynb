```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.shape
```

```python
sns.set_style('whitegrid')
```

```python
sns.displot(tips['total_bill'])
```

```python
#parametras kde gali nuimti lijiną, o parametras bins padidinti stulpelių skaičių.

sns.displot(tips['total_bill'], kde=True, bins=20)
```

```python
sns.scatterplot(tips, x='total_bill',  y='tip', size = 'size')
```

```python
sns.color_palette('rocket')
```

```python
sns.jointplot(tips, x='total_bill', y='tip', kind = "hex", palette="flare", cmap='rocket_r', color=sns.color_palette("rocket")[1])
#Matome, koks yra sąntykio tarp 'total_bill' ir 'tip' pasiskirstymas. Pagrindiniame 
#lange yra sklaidos diagrama, o iš dešinės ir viršuje - tos pačios histogramos, iš kurių sukombinavome rezultatą.
```

```python
sns.jointplot(tips, x='total_bill', y='tip', hue="sex")
```

```python
sns.jointplot(tips, x='total_bill', y='tip', hue="size", palette='coolwarm')
```

```python
sns.jointplot(tips, x='total_bill', y='tip', kind='reg', palette="coolwarm")
```

```python
sns.jointplot(tips, y='total_bill', x='tip', kind="kde", fill=True)
```

```python
sns.scatterplot(tips, x='total_bill', y='tip', hue='sex', size='size')
```

```python
#.pairplot() duoda mums visas įmanomas pasiskirstymo kombinacijas lentelėje:
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, diag_kind='hist', hue = 'sex')
```

```python
sns.barplot(tips, x='sex', y ='total_bill')
```

```python
sns.barplot(tips, x='sex', y ='total_bill', estimator='sum')
```

```python
sns.barplot(tips, x='sex', y ='total_bill', hue= "size", estimator='sum')
```

```python
sns.barplot(tips, x='sex', y ='total_bill', hue= "day", estimator='sum')
```

```python
sns.countplot(tips, x='smoker')
```

```python
sns.countplot(tips, x='size')
```

```python
sns.boxplot(tips, x="smoker", y="total_bill")
```

```python
# .heatmap()
# pandas turi metodą, kuris parodo skaičių koreliacijas lentelėje:
```

```python
koreliacijos = tips.corr(numeric_only=True)
koreliacijos
```

```python
sns.heatmap(koreliacijos, annot=True)
```

```python
g = sns.FacetGrid(tips, col='time', row='sex')
g.map(sns.scatterplot, 'tip', 'total_bill')
```

```python
g = sns.FacetGrid(data = tips, col='time', row='smoker')
g.map(sns.countplot, 'size')
```
