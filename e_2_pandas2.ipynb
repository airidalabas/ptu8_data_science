```python
import numpy as np
import pandas as pd
```

```python
top_left = pd.read_csv('top1-25-1.csv')
top_right = pd.read_csv('top1-25-2.csv')
bottom_left = pd.read_csv('top26-50-1.csv')
bottom_right = pd.read_csv('top26-50-2.csv')
```

```python
bottom_table = pd.concat([bottom_left, bottom_right], axis=1)
top_table = pd.merge(top_left, top_right, on=["Popularity", "Track.Name"])
data = pd.concat([top_table, bottom_table], sort=False)
data.head()
```

```python
data["nr"] = np.arange(1, 51)
data.set_index('nr', inplace=True)
data.head()
```

```python
#3 Sukurkite grupavimo pagal žanrą objektą
data.groupby('Genre')
```

```python
music_genre = data.groupby('Genre')
music_genre.head()
```

```python
#4 kokie žanrai lentelėje pasitaiko daugiau negu 3 kartus?
new = music_genre.count()
new
```

```python
new[new['Energy']>3]['Energy']
```

```python
#5 Koks žanras pats populiariausias? Koks mažiausiai populiarus? (pagal populiarumo vidurkį)
poppuliarumo_vidurkis_pagal_zanra = music_genre["Popularity"].mean().min()
least_popular_genre = music_genre['Popularity'].mean().idxmin()

most_popular_genre = music_genre['Popularity'].mean().idxmax()
most_popular_score = music_genre['Popularity'].mean().max()

print(most_popular_genre, most_popular_score)
print(least_popular_genre, poppuliarumo_vidurkis_pagal_zanra)
```

```python
#6 Sukurkite lentelę, kurioje matytųsi, koks žanras turi aukščiausią vidurkį kiekviename indikatoriuje, bei pats vidurkis.
lentele1 = data.columns.to_list()[3:]
lentele3 = music_genre.mean(numeric_only=True).max()
lentele2 = music_genre.mean(numeric_only=True).idxmax()

df = pd.DataFrame([lentele1, lentele2, lentele3], 
                ['Indikatorius', 'Žanras', 'Balai']).transpose()
df["nr"] = np.arange(1, 11)
df.set_index("nr", inplace=True)
df
```
